```python
import pytesseract
from PIL import Image
import csv

# Load the image
image_path = '/kaggle/input/bills-jpg/jpg.png'  # Replace with your image file path
img = Image.open(image_path)

# Extract text from the image
extracted_text = pytesseract.image_to_string(img)

# Split the extracted text into lines (this can be adjusted based on how you want to structure the data)
lines = extracted_text.split('\n')

# Write the extracted text into a CSV file
csv_file_path = '/kaggle/working/trychey.csv'  # Path to save the CSV file

with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Optional: Write a header (if needed)
    writer.writerow(["Extracted Text"])
    
    # Write each line from the extracted text into the CSV
    for line in lines:
        if line.strip():  # Avoid writing empty lines
            writer.writerow([line])

print(f"Extracted text has been saved to {csv_file_path}")
```
